```python
from electricity_price_predictor.data import get_data, get_past_weather
import pandas as pd
```

```python
df = get_data()
df[df.index.hour==10].tail(10)
```

```python
tmp = df.copy()
```

```python
from datetime import timedelta, date, datetime
day = date.today()-timedelta(days=4)
past = get_past_weather(day)
past.head()
```

```python
columns = ['temp', 'humidity', 'wind_speed']
for col in columns:
    tmp.loc[past.index[0]:, col] = past[col]
```

```python
tmp.index.name='dt'
tmp[tmp.index.hour==10].tail(10)
```

```python
tmp.drop(columns=['price', 'holiday', 'weekend'], inplace=True)
tmp.to_csv('../raw_data/updated_weather.csv')
```
